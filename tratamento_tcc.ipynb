```python
def criar_tabela(dimensao, nome):
    dim = pd.DataFrame(df[dimensao].unique())
    dim = dim.reset_index().rename(columns={'index':f'ID_{nome}', 0:f'Desc_{nome}'})
    
    return dim
```
